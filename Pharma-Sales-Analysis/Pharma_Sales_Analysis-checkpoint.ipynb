```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"Data/pharma_sales.csv"
)

df.head()
df.isnull().sum()
df.drop_duplicates(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df.info()
df.describe()
monthly_sales = df.groupby(
    df['Date'].dt.month
)['Revenue'].sum()


plt.figure(figsize=(8,5))

plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker='o'
)

plt.title("Monthly Pharma Sales Trend")

plt.xlabel("Month")

plt.ylabel("Revenue")

plt.show()
top_products = (
    df.groupby('Product_Name')
    ['Revenue']
    .sum()
    .sort_values(
        ascending=False
    )
    .head(10)
)


plt.figure(figsize=(10,5))

top_products.plot(
    kind='bar'
)

plt.title(
    "Top 10 Performing Products"
)

plt.xlabel(
    "Product Name"
)

plt.ylabel(
    "Revenue"
)

plt.xticks(rotation=45)

plt.show()
category_sales = (
    df.groupby('Category')
    ['Revenue']
    .sum()
)


plt.figure(figsize=(7,7))

category_sales.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title(
    "Sales Distribution by Category"
)

plt.ylabel("")

plt.show()
region_sales = (
    df.groupby('Region')
    ['Revenue']
    .sum()
)


region_sales.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title(
    "Revenue by Region"
)

plt.xlabel(
    "Region"
)

plt.ylabel(
    "Revenue"
)

plt.show()
```
